```python
# импортируем необходимые библиотеки
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# seaborn может выводить предупреждения о работе с определенным типом данных. Их можно отключить

# import warnings; warnings.simplefilter('ignore')
```

```python
# загружаем датасет
df = pd.read_csv('tips.xls')
```

```python
# для построения гистограммы распределения количественных признаков используется метод distplot
sns.set() # стандартные параметры отображения
sns.distplot(df['total_bill'])
```

```python
# seaborn позволяет разместить на одном графике две гистограммы и показать связь между ними
sns.jointplot(x='total_bill', y = 'tip', data = df)

#характер представления информации в центральной части можно регулировать
# kind = 'reg' выводит не только точки, но и регрессию - выражение общего характера связи
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'reg')
```

```python
# Другой способ отобразить гистограммы распределения  — построить график типа pairplot:
sns.pairplot(df)
```
