```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('_tips.xls')
```

```python
# Объектно-ориентированный подход к созданию графиков
fig = plt.figure() #создаётся объект, к которому применяе все настройки
axes = fig.add_axes([0,0,1,1]) # создаём координатную плоскость
```

```python
# далее в созданно плоскости мы можем разместить любую графическую информацию
fig = plt.figure() #создаётся объект, к которому применяе все настройки
axes = fig.add_axes([0,0,1,1]) # создаём координатную плоскость
axes.hist(df['total_bill'], bins = 100, color = 'black') # созданный объект применяем к DataFrame 
axes.set_title('Common') # добавляем наименование графика
axes.set_ylabel('vertical') # добавляем наименование оси Y
axes.set_xlabel('gorizont') # добавляем наименование оси X

# параметр bins меняет количество столбцов для демонстрации распределения 
```

```python
fig = plt.figure() #создаётся объект, к которому применяе все настройки
axes = fig.add_axes([0,0,1,1]) # создаём координатную плоскость
axes.scatter(x = df['total_bill'], y = df['tip'], marker = 'v', color = 'black') # созданный объект применяем к DataFrame 
axes.set_title('Common') # добавляем наименование графика
axes.set_ylabel('vertical') # добавляем наименование оси Y
axes.set_xlabel('gorizont') # добавляем наименование оси X

# marker позволяет настроить вид маркера, для каждого отдельного случая.
```

```python
# несколько систем координат для отображения вспомогательной информации.

fig = plt.figure()
main_axes = fig.add_axes([0,0,1,1])
auxilliary_axes = fig.add_axes([0.6,0.6,0.38,0.38]) 

#для размещения двух графиков в одном окне подготовим DataFrame. 

tips_per_day = df.groupby(by='day')['total_bill'].agg(['min','max']).reset_index()
display(tips_per_day)
main_axes.hist(df['total_bill'])
# auxilliary_axes.bar(x = tips_per_day['day'], height = tips_per_day['max'])
# auxilliary_axes.bar(x = tips_per_day['day'], height = tips_per_day['min'])

# чтоб на вспомогательной диаграме графики друг друга не перекрывали - поработаем с настройками
auxilliary_axes.bar(x = tips_per_day['day'], height = tips_per_day['max'],
                   width = 0.4, align = 'edge')
auxilliary_axes.bar(x = tips_per_day['day'], height = tips_per_day['min'],
                   width = -0.4, align = 'edge')



```
