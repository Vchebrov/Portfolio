```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(start=-3.,stop = 3., num = 1000) #создаем массив из 1000 чисел от -3 до 3
y1 = np.exp(x) # создаем переменную, как экспоненту
```

```python
# Функция matplotlib.pyplot.plot позволяет отрисовать последовательность значений.

fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y1)
```

```python
# отрисовка нескольких графиков
y2 = 10 + 0.4*x - 0.3*x**2 + 0.1*x**3

fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x[500:],y1[500:]) #в квадратных скобках указана точка отсчета
axes.plot(x,y2)   
```

```python
# стили линий

fig = plt.figure(figsize=(7,4))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y1,linestyle='--') # linestyle задаем тип линии
```

```python
# цвета и маркеры
# Чтобы задать тип маркера и цвет линии, воспользуемся методом форматирования линии.

fig=plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x[::30],y1[::30],'mD') #в квадратных скобках указан весь диапазон, но с шагом 30
```
