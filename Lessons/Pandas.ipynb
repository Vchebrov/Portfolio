```python
#                                    Группировка данных DataFrame

# Данные можно представить себе как последовательность каких-то единичных наблюдений. 
# Например, в нашем датасете единичное наблюдение — это конкретный футболист.

# каждое наблюдение имеет атрибуты: категориальные (строки) или численные (значения)

# Группировка подразумевает разделение по какому либо принаку и дальнейшее применение агрегирующей функциию,
# т.е. вычисление какого либо показателя. 

# Для примера групировки возьмем небольшой фрагмент
df = pd.read_csv('data_sf.xls')

small_df = df[df.columns[1:8]].head(25) # - сортировка с первого по восьмой столбец и вывод первых 25 строк.
# display(small_df)
# Оценим сколько футболистов относятся к каждой из национальностей

# Каждая строка датафрейма содержит информацию об отдельно взятой позиции, то посчитаем сколько раз встречается
# каждая национальность

s = small_df['Nationality'].value_counts()
s.index # - список уникальных значений

# value_counts() - функция подсчитывает для каждого значения в серии количество повторений. Возвращает серию.

# использование normalize = True, в качестве параметра для value_counts() выводит инфу в процентах от общего кол-ва
# display(small_df['Nationality'].value_counts(normalize = True))

#                                  Группировка по численному признаку

# Параметр bins используется для value_count(), чтоб распределить значения по числовым признакам.
# bins указывает на число "корзин"-диапазонов, в которые будут распределены значения из DataFrame

# Пример:

# small_df[small_df.Age > 20].Wage.value_counts(bins = 5) # - распределение по дипазонам зарплат игроков > 20 лет
s = small_df['Wage'].value_counts(bins = 4) # распределение з/п всех игроков по четырем диапазонам
display(s)
# Проверим, что за ногомячисты получают много денег
small_df.loc[(small_df['Wage'] > s.index[3].left)&(small_df['Wage'] <= s.index[3].right)]
```

```python
#                             Группировка данных. Функция groupby

#______________________________________________________________________________________________
# Общий вид группировки
# df.groupby(столбец_для_группировки)[столбцы_для_отображения].agg(['функция_1', 'функция_2'])
#______________________________________________________________________________________________


# Предположим надо найти клуб с самой высокой суммарной ЗП.
df.groupby(['Club']).groups # на вход передается столбец, по которому выполняется группировка
#  Это объект группировки, он хранит в себе информацию о том, какие строки датафрейма 
# (по индексным номерам) соответствуют определенной группе (в нашем случае определенному клубу)
# Пример: FC Köln соответствуют следующие строки: 166, 379, 899, 1399, 3415, 6904, 7052, 8892, 12171, 12376.

grouped_df = df.groupby(['Club']).sum()
# display(grouped_df) #Получился датафрейм, где все параметры, к которым можно было применить соответствующую агрегирующую функцию, 
# были просуммированы.Если точнее, были просуммированы все численные параметры

# display(grouped_df.loc['Ajax']) #отображаем данные по конкретному клубу, по индексу строки. Получаем серию.
# display((grouped_df.loc['Ajax']['Wage'])) #сумма ЗП всех игроков

#важно помнить, что суммировать надо только информативную метрику. К примеру нет смысла в сумме возрастов.

grouped_df = df.groupby(['Club'])['Wage'].sum().sort_values(ascending = False) #сотритуем по ЗП. У кого самая большая сумма ЗП
# Параметр ascending = False указывает на сортировку по убыванию
print(grouped_df)
```

```python
#                 Посчитайте среднюю (mean) и медианную (median) зарплату (Wage) футболистов из разных клубов (Club). 
#               В скольких клубах средняя и медианная зарплаты совпадают?

group = df.groupby(['Club'])['Wage'].agg(['mean', 'median'])

#  Каков максимальный размер средней зарплаты в этой группе клубов?
#  Как называется клуб, где игроки получают такую зарплату?

sorted = (group[group['median'] == group['mean']])
# display(sorted)
sorted[sorted['mean'] == 13000.0]
```

```python
# Определите максимальную силу (Strength) и баланс (Balance) 
# среди игроков клуба (Club) "FC Barcelona" из Аргентины ("Argentina"). 
# Ответ введите через точку с запятой без пробела.

df[(df.Nationality == 'Argentina')&(df.Club == 'FC Barcelona')][['Strength','Balance']].max()
```

```python
#                                           Сводные таблицы
# Сводная таблица обычно содержит строки, колонки и поля данных. В этом случае столбец — Club, строка — Nationality. 
# Нужные нам данные — это сумма зарплат (sum по колонке Wage).
# Сводные таблицы допускают несколько видов функций агрегации, включая суммирование, нахождение среднего, 
# стандартное отклонение, количество элементов и другие. 
# Всё зависит от поставленной задачи:

#                                        Функция pivot_table

pivot = df.loc[df['Club'].isin(['FC Barcelona','Real Madrid','Juventus','Manchester United'])].pivot_table(values=['Wage'],
index=['Nationality'],
columns=['Club'],
aggfunc='sum')
# позволяет отфильтровать датафрейм по колонке Club, 
# где значение в колонке соответствует одному из перечисленных в массиве клубов.
#  В параметр values мы передаем ту колонку, по которой нам нужно строить сводные данные, применяя агрегирующую функцию сумма.
#  В параметр index передадим ту колонку, данные которой будут представлены строками сводной таблицы.
#  В параметр columns передаем колонку, значения которой будут в столбцах.
#  В параметр aggfunc передаем агрегирующую функцию. 
#  Для того, чтобы добавить столбец и строку с итоговыми значениями, нужно добавить параметр margins=True:

display(pivot)

```
