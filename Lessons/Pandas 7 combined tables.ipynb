```python
import pandas as pd

ratings = pd.read_csv('ratings.xls')
movies = pd.read_csv('movies.xls')

display(ratings.head())
display(movies)
```

```python
#                                       Объединение таблиц
# у movies и ratings есть общий стоблец movieid. Значит можно собрать эти DataFrame в один методом merge

joined = ratings.merge(movies, on = 'movieId', how = 'left')
display(joined.head())
len(ratings) == len(joined)

#Схематично метод merge можно описать так: joined = left_df.merge(right_df, on='', how='').
# left_df/right_df это таблицы, которые объединяем. К правому присоединяем левый
# on - определяет стоблец по которому происходит объединение. Если по нескольким, то ['col_1', 'col_2']


# how — параметр объединения записей. Он может иметь четыре значения: left, right, inner и outer. 
# При значении left берем все записи (movieId) из "левого" датафрейма (ratings) и ищем их соответствия в "правом" (movies). 
# В итоговом датафрейме останутся только те значения, которым были найдены соответствия, то есть только значения из ratings. 
# Аналогично при параметре right остаются только значения из "правого" датафрейма. 
# Если совпадений между таблицами нет, то ставим нулевое значение. 
# Значение inner оставляет только те записи (movieId), которые есть в обоих датафреймах, 
# outer объединяет все варианты movieId в обоих датафреймах. 

# joined = ratings.merge(movies, on = 'movieId', how = 'right')
# display(joined.head())
# len(ratings) == len(joined)

# joined = ratings.merge(movies, on = 'movieId', how = 'inner')
# display(joined.head())
# # len(ratings) == len(joined)

# joined = ratings.merge(movies, on = 'movieId', how = 'outer')
# display(joined.head())

# # len(ratings) == len(joined)
```

```python
ratings = pd.read_csv('ratings_example.txt', sep = '\t')
movies = pd.read_csv('movies_example.txt', sep = '\t')
display(ratings, movies)

joined = ratings.merge(movies, on = 'movieId', how = 'right')
display('right', joined)

joined = ratings.merge(movies, on = 'movieId', how = 'left')
display('left', joined)

joined = ratings.merge(movies, on = 'movieId', how = 'inner')
display('inner', joined)

joined = ratings.merge(movies, on = 'movieId', how = 'outer')
display('outer', joined)
```

```python
#                              Удаляем дубликаты строк после объединения
# метод drop_duplicates
movies.drop_duplicates(subset = 'movieId', keep = 'first', inplace = True)
display(movies)

# subset - столбец, где ищем дубликаты. Можно указать несколько
# keep - какой оставляем?
# inplace - указывает, где сохраняем изменения. В том же DataFrame или нет
new = ratings.merge(movies, on = 'movieId', how = 'left')
display('new', new)
```
